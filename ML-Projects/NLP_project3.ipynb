```python
df['Label']=df['Label'].map({'Non-Spam':0,'Spam':1})
df
```

```python
sms=df.Message_body
sms
```

```python
from nltk import TweetTokenizer
tk=TweetTokenizer()
sms=sms.apply(lambda x:tk.tokenize(x)).apply(lambda x:' '.join(x))
sms
```

```python
sms=sms.str.replace('[^a-zA-Z0-9]+',' ')
sms
```

```python
from nltk.tokenize import word_tokenize
sms=sms.apply(lambda x:' '.join([w for w in word_tokenize(x) if len(w)>=3]))
sms
```

```python
from nltk.stem import SnowballStemmer
st=SnowballStemmer('english')
sms=sms.apply(lambda x:[st.stem(i.lower()) for i in tk.tokenize(x)]).apply(lambda x:' '.join(x))
sms
```

```python
from nltk.corpus import stopwords
sw=stopwords.words('english')
sms=sms.apply(lambda x:[i for i in word_tokenize(x) if i not in sw]).apply(lambda x:' '.join(x))
sms
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vec=TfidfVectorizer()
data=vec.fit_transform(sms)
data
```

```python
print(data)
```

```python
y=df['Label'].values
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(data,y,test_size=0.30,random_state=42)
x_train
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import BernoulliNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
knn=KNeighborsClassifier(n_neighbors=7)
model=SVC()
base=BernoulliNB()
tree=DecisionTreeClassifier(criterion='entropy')
rf=RandomForestClassifier(random_state=100)
lst=[knn,model,base,tree,rf]
```

```python
for i in lst:
  i.fit(x_train,y_train)
  y_pred=i.predict(x_test)
  print('Model is:',i)
  print(y_pred)
  print(accuracy_score(y_test,y_pred))
  print(confusion_matrix(y_test,y_pred))
  print(classification_report(y_test,y_pred))
  print('**'*100)
```
